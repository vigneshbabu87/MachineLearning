### 1.We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below. 
 
Task: 
 
1. fitting it to the periodic function 
 
2. plot the fit 
 
Data 
 
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25 
 
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18 

```python
import numpy as np

max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
import matplotlib.pyplot as plt

months = np.arange(12)
plt.figure(figsize=(8,6))

#red circle "ro" - indicates maximum temperature
plt.plot(months, max_temp, 'ro')

#Blue circle "bo" - indicates minimum temperature
plt.plot(months, min_temp, 'bo')

plt.xlabel('Months')
plt.ylabel('Min and max temperature')
```

```python
###Fitting it to the periodic function

from scipy import optimize

def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months, max_temp, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months, min_temp, [-40, 20, 0])
```

```python
#Plot the fit

days = np.linspace(0, 12, num=365)

plt.figure(figsize=(8,6))
plt.plot(months, max_temp, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, min_temp, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Months')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

Charts to plot:  
1. Create a pie chart presenting the male/female proportion 
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender 


```python
import pandas as pd

url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```python
#1. Create a pie chart presenting the male/female proportion
Gender = titanic['sex'].astype('category')
Gender = Gender.cat.codes
Gender.head()
```

```python
plt.pie(Gender.value_counts(), labels = titanic['sex'].unique(), autopct='%1.1f%%')
plt.title('Male/Female Chart')
plt.axis('equal')
plt.show()
```

```python
#Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
plt.figure(figsize=(10,8))
plt.scatter(titanic['age'], titanic['fare'], c = Gender)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
